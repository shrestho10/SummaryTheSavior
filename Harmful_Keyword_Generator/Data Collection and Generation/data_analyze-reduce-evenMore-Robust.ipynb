```python
import numpy as np
import pandas as pd
from collections import Counter

# Parameters
num_prompts = len(remaining_df)
num_queries = len(remaining_df_benign_queries)
queries_per_prompt = 50
min_queries_per_query = 50

# Create random prompts and queries with additional columns
benign_prompts = remaining_df
benign_queries = remaining_df_benign_queries
# Calculate total pairs needed
total_pairs = num_prompts * queries_per_prompt
min_total_queries_needed = num_queries * min_queries_per_query

# Ensure we have enough total pairs
if total_pairs < min_total_queries_needed:
    raise ValueError("Total pairs are not enough to ensure each query appears at least 20 times.")

# Initialize tracking structures
all_pairs = []
query_count = Counter()
prompt_count = Counter()

# Iterate over each prompt
for prompt in benign_prompts['prompt']:
    # Filter queries that haven't reached the minimum appearance count
    remaining_queries = benign_queries['query'][benign_queries['query'].apply(lambda q: query_count[q] < min_queries_per_query)].tolist()
    
    # If not enough remaining queries, use all queries
    if len(remaining_queries) < queries_per_prompt:
        remaining_queries = benign_queries['query'].tolist()
    
    # Randomly sample queries for the current prompt
    queries_for_prompt = np.random.choice(remaining_queries, size=queries_per_prompt, replace=False)
    
    # Update query counts, prompt counts and add pairs
    for query in queries_for_prompt:
        query_count[query] += 1
        prompt_count[prompt] += 1
        all_pairs.append({'prompt': prompt, 'query': query})

# Convert to DataFrame
dataset_df = pd.DataFrame(all_pairs)

# Merge additional columns
merged_df = dataset_df.merge(benign_prompts, on='prompt', how='left')
merged_df = merged_df.merge(benign_queries, on='query', how='left')



# Calculate occurrences
query_occurrences = pd.Series(query_count).sort_values(ascending=False)
prompt_occurrences = pd.Series(prompt_count).sort_values(ascending=False)

# Print details
print(f"Total pairs: {len(merged_df)}")
print(f"Unique prompts: {merged_df['prompt'].nunique()}")
print(f"Unique queries: {merged_df['query'].nunique()}")
print(f"Minimum occurrences of each query: {min(query_count.values())}")

```

```python
import pandas as pd


new_df = pd.DataFrame()

def clean_text(text):
    if pd.isna(text) or text is None:
        return ''
    
    # Remove unwanted characters
    if isinstance(text, str):
        text = text.replace('[', '').replace(']', '').replace(', ', ',').strip()
    
    # Convert to lowercase for consistent comparison
    text = text.lower()
    
    # Return empty string if text matches any of the unwanted phrases
    if text in ['none', 'no keywords', 'no keywords found', 'no keywords identified', '','no harmful keywords', 'no harmful keywords found', 'no harmful keywords identified', 'n/a']:
        return ''
    
    return text
    
def extract_summary(text):
    # Extract part after 'summary:' if it exists
    if isinstance(text, str) and 'summary:' in text.lower():
        return text.lower().split('summary:', 1)[1].strip()
    return text

rows=[]
counter=0

data1=merged_df


for i in range(0, len(data1)):
    counter=counter+1
    if(counter%10000==0):
        print(counter)
    text1 = data1['prompt'][i]
    text2 = data1['query'][i]
    total = text1 + ". Also " + text2
    text3 = data1['harmful_keywords_x'][i]
    text3=clean_text(text3)
    text4 = data1['harmful_keywords_y'][i]
    text4=clean_text(text4)
    if text3 == '' and text4 =='':
        total2 = ''
    elif text3 == '' and text4 !='':
        total2 = text4
    elif text3 != '' and text4 =='':
        total2 = text3       
    else:
        punc = ','
        total2 = text3 + punc + text4
    text5 = data1['summary_x'].iloc[i]
    text6 = data1['summary_y'].iloc[i]
    total3 = extract_summary(text5) + ". Also in the last sentence " + extract_summary(text6)
    rows.append({'prompt': total, 'harmful_keywords': total2, 'summary': total3})


new_df = pd.DataFrame(rows)
new_df
```

```python
import pandas as pd


new_df = pd.DataFrame()

def clean_text(text):
    if pd.isna(text) or text is None:
        return ''
    
    # Remove unwanted characters
    if isinstance(text, str):
        text = text.replace('[', '').replace(']', '').replace(', ', ',').strip()
    
    # Convert to lowercase for consistent comparison
    text = text.lower()
    
    # Return empty string if text matches any of the unwanted phrases
    if text in ['none', 'no keywords', 'no keywords found', 'no keywords identified', '','no harmful keywords', 'no harmful keywords found', 'no harmful keywords identified', 'n/a']:
        return ''
    
    return text
    
def extract_summary(text):
    # Extract part after 'summary:' if it exists
    if isinstance(text, str) and 'summary:' in text.lower():
        return text.lower().split('summary:', 1)[1].strip()
    return text

rows=[]
counter=0

data1=merged_df


for i in range(0, len(data1)):
    counter=counter+1
    if(counter%10000==0):
        print(counter)
    text1 = data1['prompt'][i]
    text2 = data1['query'][i]
    total = text1 + ". Also " + text2
    text3 = data1['harmful_keywords_x'][i]
    text3=clean_text(text3)
    text4 = data1['harmful_keywords_y'][i]
    text4=clean_text(text4)
    if text3 == '' and text4 =='':
        total2 = ''
    elif text3 == '' and text4 !='':
        total2 = text4
    elif text3 != '' and text4 =='':
        total2 = text3       
    else:
        punc = ','
        total2 = text3 + punc + text4
    text5 = data1['summary_x'].iloc[i]
    text6 = data1['summary_y'].iloc[i]
    total3 = extract_summary(text5) + ". Also in the last sentence " + extract_summary(text6)
    rows.append({'prompt': total, 'harmful_keywords': total2, 'summary': total3})


new_df = pd.DataFrame(rows)
new_df
```
