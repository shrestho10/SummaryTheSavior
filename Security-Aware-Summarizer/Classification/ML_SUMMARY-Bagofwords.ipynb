```python
import re

def remove_special_characters(text):
    return re.sub(r'[^\w\s]', '', text)  # Remove punctuation

def to_lowercase(text):
    return text.lower()

def remove_extra_whitespace(text):
    return ' '.join(text.split())

from nltk.corpus import stopwords

def remove_stop_words(text):
    stop_words = set(stopwords.words('english'))
    words = text.split()
    return ' '.join(word for word in words if word not in stop_words)

def remove_extra_words(text):
    stop_words1 = ['prompt','intent','intention']
    words = text.split()
    return ' '.join(word for word in words if word not in stop_words1)

from nltk.tokenize import word_tokenize

def tokenize_text(text):
    return word_tokenize(text)

from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

# Stemming
stemmer = PorterStemmer()
def stem_words(words):
    return [stemmer.stem(word) for word in words]

# Lemmatization
lemmatizer = WordNetLemmatizer()
def lemmatize_words(words):
    return [lemmatizer.lemmatize(word) for word in words]

def remove_duplicates(text):
    seen = set()
    result = []
    for line in text.split('\n'):
        if line not in seen:
            seen.add(line)
            result.append(line)
    return '\n'.join(result)

def normalize_text(text):
    # Example: Normalize date formats
    text = re.sub(r'\d{2}/\d{2}/\d{4}', 'DATE', text)
    return text
```

```python
def clean_document(text):
    text = remove_special_characters(text)
    text = to_lowercase(text)
    text = remove_extra_whitespace(text)
    text = remove_extra_words(text)
    text = remove_stop_words(text)

    tokens = tokenize_text(text)
    tokens = lemmatize_words(tokens)  # Use stemming if preferred
    cleaned_text = ' '.join(tokens)
    return cleaned_text
```
