```python
d4=pd.read_csv("./wildjailbreak_part1_llama3.1_70B_final_last_attempt.csv")
d4["Llama_classification_1"] = d4["Llama_classification_1"].str.split("classification").str[1].str.strip()
d4=d4[['prompt','Llama_summary_1','Llama_classification_1','harmful_keywords']]
d4
```
