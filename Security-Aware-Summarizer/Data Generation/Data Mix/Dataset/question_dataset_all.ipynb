```python
import pandas as pd
```

```python
d1=pd.read_csv("./dan_ques_llama3.1_70B_final_last_attempt.csv")
d1["Llama_classification_1"] = d1["Llama_classification_1"].str.split("classification").str[1].str.strip()
d1=d1[['prompt','Llama_summary_1','Llama_classification_1','harmful_keywords']]
d1
```

```python
d2=pd.read_csv("./other_ques_llama3.1_70B_final_last_Attempt.csv")
d2["Llama_classification_1"] = d2["Llama_classification_1"].str.split("classification").str[1].str.strip()
d2=d2[['prompt','Llama_summary_1','Llama_classification_1','harmful_keywords']]
d2
```

```python
d3=pd.read_csv("./wildjailbreak_part2_llama3.1_70B_final_last_attempt.csv")
d3["Llama_classification_1"] = d3["Llama_classification_1"].str.split("classification").str[1].str.strip()
d3=d3[['prompt','Llama_summary_1','Llama_classification_1','harmful_keywords']]
d3=d3[2000:4000]
d3
```

```python
d4=pd.read_csv("./or_bench_6000_summary_last_attempt.csv")
d4["Llama_classification_1"] = d4["Llama_classification_1"].str.split("classification").str[1].str.strip()
d4=d4[['prompt','Llama_summary_1','Llama_classification_1','harmful_keywords']]
d4
```

```python
d5=pd.read_csv("./or_bench_5000_summary_part2_last_attempt.csv")
d5["Llama_classification_1"] = d5["Llama_classification_1"].str.split("classification").str[1].str.strip()
d5=d5[['prompt','Llama_summary_1','Llama_classification_1','harmful_keywords']]
d5
```

```python
ques_all=pd.concat([d1,d2,d3,d4,d5],axis=0,ignore_index=True)
ques_all
```
