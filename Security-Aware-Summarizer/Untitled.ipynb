```python
import nltk
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt
from collections import Counter

# Ensure you have downloaded the punkt tokenizer
nltk.download('punkt')

# Example assuming the column containing the text is 'text_column'
# Replace 'text_column' with the actual name of the text column in your CSV file
text_data = d1['text_column'].dropna()  # Drop any NaN values

# Tokenize all the text data
tokens = text_data.apply(word_tokenize).explode()

# Count the frequency of each token
token_counts = Counter(tokens)

# Show the most common tokens
print(token_counts.most_common(10))
```
