```python
import pandas as pd

# Example DataFrames
df1 = pd.read_csv("./jbc_gpt4.csv")
df2 = pd.read_csv("./jbc_llama2.csv")
df3 = pd.read_csv("./jbc_vicuna.csv")

# Normalize prompts
def normalize_prompts(df):
    return df['prompt']

prompts1 = set(normalize_prompts(df1))
prompts2 = set(normalize_prompts(df2))
prompts3 = set(normalize_prompts(df3))


# Compare all pairs
same_prompts = (prompts1 == prompts2 == prompts3)

# Print result
if same_prompts:
    print("All datasets have the same prompts.")
else:
    print("The datasets have different prompts.")


```
